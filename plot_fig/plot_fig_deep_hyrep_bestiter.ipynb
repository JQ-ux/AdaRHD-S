```python
import pickle
import numpy as np

ratio = 0.05
with open(f'../results/hyrep_spd_AdaRHD_cg_lr0.1data_ratio{ratio}.pkl', 'rb') as handle:
    data = pickle.load(handle)

# Get the accuracy data from the data dictionary
all_accuracies = data['accuracy']
new_all_accuracies = []
for i in range(len(all_accuracies)):
    new_all_accuracies.append(all_accuracies[i].cummax(dim=1)[0])
all_accuracies = new_all_accuracies

all_test_accuracies = data['test_accuracy']
new_all_test_accuracies = []
for i in range(len(all_test_accuracies)):
    new_all_test_accuracies.append(all_test_accuracies[i].cummax(dim=1)[0])
all_test_accuracies = new_all_test_accuracies

# Determine how many experiments we have
num_experiments = len(all_accuracies)
print(f"Found {num_experiments} experiments in the data")

# Prepare arrays for storing padded data
train_accuracies = []
test_accuracies = []
max_epochs = 300  # Target length for all experiments

# Process each experiment
for i in range(num_experiments):
    # Extract accuracy data and reshape
    train_acc = all_accuracies[i].reshape(-1)
    
    # Get the number of epochs in this experiment
    num_epochs = len(train_acc)
    print(f"Experiment {i+1} has {num_epochs} epochs")
    
    # If less than max_epochs, pad with the last value
    if num_epochs < max_epochs:
        train_acc = np.pad(train_acc, (0, max_epochs - num_epochs), 
                            'constant', constant_values=train_acc[-1])
    else:
        # If more than max_epochs, truncate
        train_acc = train_acc[:max_epochs]
        train_acc = train_acc.numpy()
    
    # Add to our list
    train_accuracies.append(train_acc)
    
    # Same for test accuracy if available
    if i < len(all_test_accuracies):
        test_acc = all_test_accuracies[i].reshape(-1)
        if len(test_acc) < max_epochs:
            test_acc = np.pad(test_acc, (0, max_epochs - len(test_acc)),
                                'constant', constant_values=test_acc[-1])
        else:
            test_acc = test_acc[:max_epochs]
            test_acc = test_acc.numpy()
        test_accuracies.append(test_acc)

# Convert lists to numpy arrays for easier manipulation
train_accuracies1 = np.array(train_accuracies)
test_accuracies1 = np.array(test_accuracies) if test_accuracies else None
```

```python
import pickle
import numpy as np
import os
import torch

# --- 实验配置 (针对 lr=0.1) ---
lr = 0.1
max_epochs = 3000  # 根据你记录的数据：lr=0.1 对应 3000 epochs
method = 'AdaRHD_S'
ratio = 0.05
results_dir = r"C:\Users\24300\Desktop\Ada_S\results"

# 拼接具体的文件路径
file_path = os.path.join(results_dir, f'hyrep_spd_{method}_lr{lr}_data_ratio{ratio}.pkl')

with open(file_path, 'rb') as handle:
    data = pickle.load(handle)

# --- 1. 提取并处理 Accuracy (执行 cummax 以保证曲线单调递增) ---
all_accuracies = data['accuracy']
new_all_accuracies = []
for i in range(len(all_accuracies)):
    # 模仿示例：使用 dim=1 并获取值 [0]
    new_all_accuracies.append(all_accuracies[i].cummax(dim=1)[0])
all_accuracies = new_all_accuracies

all_test_accuracies = data['test_accuracy']
new_all_test_accuracies = []
for i in range(len(all_test_accuracies)):
    new_all_test_accuracies.append(all_test_accuracies[i].cummax(dim=1)[0])
all_test_accuracies = new_all_test_accuracies

# Determine how many experiments we have
num_experiments = len(all_accuracies)
print(f"Found {num_experiments} experiments in the data")

# Prepare arrays for storing padded data
train_accuracies = []
test_accuracies = []

# --- 2. 循环处理每组实验数据 (Padding/Truncate) ---
for i in range(num_experiments):
    # Extract accuracy data and reshape
    train_acc = all_accuracies[i].reshape(-1)
    
    # Get the number of epochs in this experiment
    num_epochs = len(train_acc)
    print(f"Experiment {i+1} has {num_epochs} epochs")
    
    # If less than max_epochs, pad with the last value
    if num_epochs < max_epochs:
        # 使用 .cpu().numpy() 确保跨平台兼容
        train_acc = np.pad(train_acc.cpu().numpy(), (0, max_epochs - num_epochs), 
                            'constant', constant_values=train_acc[-1].item())
    else:
        # If more than max_epochs, truncate
        train_acc = train_acc[:max_epochs].cpu().numpy()
    
    # Add to our list
    train_accuracies.append(train_acc)
    
    # Same for test accuracy if available
    if i < len(all_test_accuracies):
        test_acc = all_test_accuracies[i].reshape(-1)
        if len(test_acc) < max_epochs:
            test_acc = np.pad(test_acc.cpu().numpy(), (0, max_epochs - len(test_acc)),
                                'constant', constant_values=test_acc[-1].item())
        else:
            test_acc = test_acc[:max_epochs].cpu().numpy()
        test_accuracies.append(test_acc)

# --- 3. 转换成最终的 Numpy 矩阵 (序号为 4) ---
train_accuracies5 = np.array(train_accuracies)
test_accuracies5 = np.array(test_accuracies) if test_accuracies else None

print(f"处理完成。lr={lr} 的矩阵形状: {train_accuracies4.shape}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# --- 1. 均值与标准差计算 ---
# 请根据你实际加载的变量名对应以下 5 个算法
# 假设：1:CG, 2:GD, 3:RHGD-50, 4:AdaRHD-S, 5:RHGD-20 (请根据实际情况微调)
plot_configs = [
    (train_accuracies4, 'red',    'AdaRHD-CG'),
    (train_accuracies1, 'blue',   'AdaRHD-GD'),
    (train_accuracies2, 'green',  'RHGD-50'),
    (train_accuracies3, 'orange', 'RHGD-20'),
    (train_accuracies5, 'cyan',   'AdaRHD-S')  # 假设第5组是 RHGD-20
]

# --- 2. 绘图设置 ---
os.makedirs('../fig/', exist_ok=True)
plt.rcParams['axes.grid'] = True

def plot_and_save(xlim_range, suffix):
    plt.figure(figsize=(6, 5))
    
    for data_matrix, color, label in plot_configs:
        # 计算统计量
        mean_acc = np.mean(data_matrix, axis=0)
        std_acc = np.std(data_matrix, axis=0)
        epochs = range(len(mean_acc))
        
        # 绘图
        plt.plot(epochs, mean_acc, color=color, label=label, linewidth=1.5)
        # 填充标准差区域 (0.5倍标准差使图面更整洁)
        plt.fill_between(epochs, 
                         mean_acc - 0.5 * std_acc, 
                         mean_acc + 0.5 * std_acc, 
                         color=color, alpha=0.2)

    plt.xlabel('Epochs', fontsize=17)
    plt.ylabel('Validation Accuracy', fontsize=17)
    plt.xlim(xlim_range)
    plt.ylim([0, 1.02])
    plt.legend(loc='lower right', fontsize=12, frameon=True, shadow=True)
    plt.grid(linestyle='--', alpha=0.5)
    plt.tight_layout()
    
    save_path = f'../fig/comparison_{suffix}.pdf'
    plt.savefig(save_path, format='pdf', bbox_inches='tight')
    plt.show()
    print(f"已保存对比图至: {save_path}")

# --- 3. 执行绘制 ---

# 图 A: 全过程展示 (3000 Epochs)
plot_and_save([-50, 3050], "full_3000")

# 图 B: 早期收敛展示 (前 100 Epochs)
plot_and_save([-2, 102], "zoom_100")

# 图 C: 中期收敛展示 (前 300 Epochs，老师有时也看这个)
plot_and_save([-5, 305], "zoom_300")
```
