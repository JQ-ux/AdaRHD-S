```python
import pickle
import numpy as np
import os
import torch

# --- 1. Configuration ---
ratio = 0.05
method = 'AdaRHD_S'
lr_list = [1.0, 0.5, 0.125, 0.1, 0.05]
base_path = r"C:\Users\24300\Desktop\Ada_S\results"

# Storage initialization (following your specific naming convention)
AdaRHD_S_val50times_means, AdaRHD_S_val50times_stds = [], []
AdaRHD_S_val70times_means, AdaRHD_S_val70times_stds = [], []
AdaRHD_S_val85times_means, AdaRHD_S_val85times_stds = [], []

print(f"开始统计方法: {method}")
print("-" * 50)

for lr in lr_list:
    # Construct file path
    file_name = f'hyrep_spd_{method}_lr{lr}_data_ratio{ratio}.pkl'
    file_path = os.path.join(base_path, file_name)
    
    if not os.path.exists(file_path):
        print(f"FileNotFound: {file_path}")
        continue
        
    with open(file_path, 'rb') as handle:
        data = pickle.load(handle)

    # In your data, these are lists of 5 tensors
    all_accs = data['accuracy']
    all_runtimes = data['runtime']
    
    num_exps = len(all_accs) # This will be 5
    v50, v70, v85 = [], [], []

    # --- Process each of the 5 runs ---
    for i in range(num_exps):
        # Flatten tensors to 1D arrays
        acc_vec = all_accs[i].reshape(-1).cpu().numpy()
        time_vec = np.cumsum(all_runtimes[i].reshape(-1).cpu().numpy())
        
        # Calculate first hitting time for each threshold
        idx50 = np.where(acc_vec >= 0.50)[0]
        if len(idx50) > 0: v50.append(time_vec[idx50[0]])
        
        idx70 = np.where(acc_vec >= 0.70)[0]
        if len(idx70) > 0: v70.append(time_vec[idx70[0]])
        
        idx85 = np.where(acc_vec >= 0.85)[0]
        if len(idx85) > 0: v85.append(time_vec[idx85[0]])

    # --- Statistical calculation and Fixed Output Format ---
    print(f"Method: {method}, Learning Rate: {lr}, Ratio: {ratio}")

    # 50% Threshold
    if v50:
        m, s = np.mean(v50), np.std(v50)
        AdaRHD_S_val50times_means.append(m)
        AdaRHD_S_val50times_stds.append(s)
        print(f"Mean time for 50% val acc: {m:.2f} ± {s:.2f} seconds")
    else:
        print("Mean time for 50% val acc: Not Reached")

    # 70% Threshold
    if v70:
        m, s = np.mean(v70), np.std(v70)
        AdaRHD_S_val70times_means.append(m)
        AdaRHD_S_val70times_stds.append(s)
        print(f"Mean time for 70% val acc: {m:.2f} ± {s:.2f} seconds")
    else:
        print("Mean time for 70% val acc: Not Reached")

    # 85% Threshold
    if v85:
        m, s = np.mean(v85), np.std(v85)
        AdaRHD_S_val85times_means.append(m)
        AdaRHD_S_val85times_stds.append(s)
        print(f"Mean time for 85% val acc: {m:.2f} ± {s:.2f} seconds")
    else:
        print("Mean time for 85% val acc: Not Reached")
            
    print("-" * 50)
```

```python
# 1. 定义全局展示的学习率列表（取并集）
# 包含：5.0, 1.0, 0.5, 0.125, 0.1, 0.05
all_lrs = [5.0, 1.0, 0.5, 0.125, 0.1, 0.05]

# 2. 各算法数据在内存中对应的索引坐标（必须与你加载数据时的顺序一致）
# 假设 CG/GD/RHGD 使用的是 [5.0, 1.0, 0.5, 0.1, 0.05]
# 假设 AdaRHD-S 使用的是 [1.0, 0.5, 0.125, 0.1, 0.05]
cg_gd_rhgd_lrs = [5.0, 1.0, 0.5, 0.1, 0.05]
s_lrs          = [1.0, 0.5, 0.125, 0.1, 0.05]

# --- LaTeX 表格头 (5个算法，每个3列，共15列数据 + 1列LR) ---
latex_code = "\\begin{table}[h]\n"
latex_code += "\\scriptsize\n"
latex_code += "\\setlength{\\tabcolsep}{0.2pt}\n" # 压缩列间距以适应页面
latex_code += "\\centering\n"
latex_code += "\\caption{Time to reach validation accuracy thresholds (seconds) for all methods}\n"
# 列格式：c|ccc|ccc|ccc|ccc|ccc
latex_code += "\\begin{tabular}{c|ccc|ccc|ccc|ccc|ccc}\n\\hline\n"
latex_code += " & \\multicolumn{3}{c|}{AdaRHD-CG} & \\multicolumn{3}{c|}{AdaRHD-GD} & \\multicolumn{3}{c|}{RHGD-50} & \\multicolumn{3}{c|}{RHGD-20} & \\multicolumn{3}{c}{AdaRHD-S} \\\\\n"
latex_code += "Step Size & 50\\% & 70\\% & 85\\% & 50\\% & 70\\% & 85\\% & 50\\% & 70\\% & 85\\% & 50\\% & 70\\% & 85\\% & 50\\% & 70\\% & 85\\% \\\\\n\\hline\n"

# 辅助函数：安全获取均值和标准差
def format_val(val_list, idx):
    if idx is not None and idx < len(val_list) and val_list[idx] is not None:
        return f"{val_list[idx]:.2f}"
    return "/"

def format_std(std_list, idx):
    if idx is not None and idx < len(std_list) and std_list[idx] is not None:
        return f"$({std_list[idx]:.2f})$"
    return "/"

# --- 填充数据行 ---
for lr in all_lrs:
    # 动态计算当前 LR 在各算法数组中的索引
    k_ref = cg_gd_rhgd_lrs.index(lr) if lr in cg_gd_rhgd_lrs else None
    k_s   = s_lrs.index(lr) if lr in s_lrs else None

    # 第一行：均值 (Mean)
    row_m = f"\\multirow{{2}}{{*}}{{{lr}}} & "
    # CG
    row_m += f"{format_val(AdaRHD_cg_val50times_means, k_ref)} & {format_val(AdaRHD_cg_val70times_means, k_ref)} & {format_val(AdaRHD_cg_val85times_means, k_ref)} & "
    # GD
    row_m += f"{format_val(AdaRHD_gd_val50times_means, k_ref)} & {format_val(AdaRHD_gd_val70times_means, k_ref)} & {format_val(AdaRHD_gd_val85times_means, k_ref)} & "
    # RHGD-50
    row_m += f"{format_val(RHGD_50_val50times_means, k_ref)} & {format_val(RHGD_50_val70times_means, k_ref)} & {format_val(RHGD_50_val85times_means, k_ref)} & "
    # RHGD-20
    row_m += f"{format_val(RHGD_20_val50times_means, k_ref)} & {format_val(RHGD_20_val70times_means, k_ref)} & {format_val(RHGD_20_val85times_means, k_ref)} & "
    # AdaRHD-S
    row_m += f"{format_val(AdaRHD_S_val50times_means, k_s)} & {format_val(AdaRHD_S_val70times_means, k_s)} & {format_val(AdaRHD_S_val85times_means, k_s)} \\\\\n"

    # 第二行：标准差 (Std)
    row_s = " & "
    row_s += f"{format_std(AdaRHD_cg_val50times_stds, k_ref)} & {format_std(AdaRHD_cg_val70times_stds, k_ref)} & {format_std(AdaRHD_cg_val85times_stds, k_ref)} & "
    row_s += f"{format_std(AdaRHD_gd_val50times_stds, k_ref)} & {format_std(AdaRHD_gd_val70times_stds, k_ref)} & {format_std(AdaRHD_gd_val85times_stds, k_ref)} & "
    row_s += f"{format_std(RHGD_50_val50times_stds, k_ref)} & {format_std(RHGD_50_val70times_stds, k_ref)} & {format_std(RHGD_50_val85times_stds, k_ref)} & "
    row_s += f"{format_std(RHGD_20_val50times_stds, k_ref)} & {format_std(RHGD_20_val70times_stds, k_ref)} & {format_std(RHGD_20_val85times_stds, k_ref)} & "
    row_s += f"{format_std(AdaRHD_S_val50times_stds, k_s)} & {format_std(AdaRHD_S_val70times_stds, k_s)} & {format_std(AdaRHD_S_val85times_stds, k_s)} \\\\\n\\hline\n"

    latex_code += row_m + row_s

# --- 表格尾 ---
latex_code += "\\end{tabular}\n\\label{tab:full_comparison}\n\\end{table}"

print(latex_code)
```
